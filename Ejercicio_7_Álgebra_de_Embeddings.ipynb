Comparación semántica

```python
y = ['coach' , 'tutor', 'trainer', 'educator' , 'professor', 'mentor', 'dog', 'tree', 'cat', 'horse', 'movie']
V_y= [vec(word) for word in y]
V_y
```

```python
from sklearn.metrics.pairwise import cosine_similarity

sim = [cosine_similarity (X=vec_word.reshape(1, -1), Y=V_x.reshape(1,-1)) for vec_word in V_y]
```

```python
z = zip(y, sim)
print(sorted(z, key=lambda x: x[1], reverse=True))
```

```python
dif = V_y[3] - V_x
e1 = dif / np.linalg.norm(dif)
```

```python
v_tmp = V_y[7] - V_x
proj_on_e1 = np.dot(v_tmp, e1) * e1
e2 = v_tmp - proj_on_e1
e2 /= np.linalg.norm(e2)
```

```python
def project(v):
  return np.array([np.dot(v, e1), np.dot(v, e2)])
```

```python
points = {
    'teacher' : vec('teacher'),
    'coach' : vec('coach'),
    'trainer' : vec('trainer'),
    'movie' : vec('movie'),
    'cat' : vec('cat'),
    'educator' : vec('educator'),
    'tutor' : vec('tutor'),
    'profesor' : vec('profesor'),
    'dog' : vec('dog'),
    'tree' : vec('tree'),
    'cat' : vec('cat'),
    'horse' : vec('horse')

}
```

```python
plt.figure(figsize=(7, 7))
for word, vector in points.items():
  x, y = project(vector)
  plt.scatter(x, y)
  plt.text(x + 0.01, y + 0.01, word)

plt.axhline(0, color = 'gray', lw= 0.5)
plt.axvline(0, color = 'gray', lw= 0.5)
plt.grid()
plt.axis('equal')
plt.show()
```
